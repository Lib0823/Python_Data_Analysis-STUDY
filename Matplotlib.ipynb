#### 프롬프트
> matplotlib.pyplot을 사용해 두 개의 세로 방향 서브플롯을 그리고 각각 그래프를 그리며, 서브플롯에 제목을 왼쪽 정렬로 표시하고 전체 그래프에는 슈퍼 타이틀을 설정하는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 사용해 x = [1, 2, 3, 4, 5], y1 = [2, 4, 6, 8, 10], y2 = [1, 2, 1, 2, 1] 데이터를 가지고 2행 1열 구조의 서브플롯을 만들고, 첫 번째 서브플롯에 y1 데이터를 선 그래프로 그리고 제목은 'Graph Sub Title 1'을 왼쪽 정렬로, 두 번째 서브플롯에 y2 데이터를 선 그래프로 그리고 제목은 'Graph Sub Title 2'를 왼쪽 정렬로 설정해줘. 전체 그래프에는 'Graph Super Title'이라는 슈퍼 타이틀을 지정하고, tight_layout()으로 서브플롯 간격을 자동 조절한 후 그래프를 표시하는 코드를 작성해줘.

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

# 첫 번째 서브 그래프
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Graph Sub Title 1', loc = 'left')

# 두 번째 서브 그래프
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Graph Sub Title 2', loc = 'left')

# 슈퍼 타이틀 설정
plt.suptitle('Graph Super Title')

# tight_layout() 함수는 서브 그래프 간의 간격을 자동으로 조절
plt.tight_layout()
plt.show()
```

#### 프롬프트
> matplotlib.pyplot을 사용해서 카테고리별 막대 그래프를 그리고, 각 막대 위에 해당 값을 텍스트로 표시하는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 불러와서 categories = ['A', 'B', 'C', 'D'], values = [25, 40, 30, 20] 데이터를 기반으로 하늘색 막대 그래프를 그리고, 각 막대 꼭대기에 값 텍스트를 가운데 정렬(ha='center'), 아래쪽 정렬(va='bottom')로 표시하며 검은색, 글자 크기 10으로 설정하는 코드를 작성해줘. 그래프 제목은 'Bar Chart with Text Annotations', x축 레이블은 'Categories', y축 레이블은 'Values'로 지정하고 그래프를 화면에 표시해줘.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]

# 막대 그래프 생성
plt.bar(categories, values, color='skyblue')

# 각 막대 머리 위치에 텍스트 추가
for i, value in enumerate(values):
  # text() 함수의 두 번째 매개변수의 값을 바꿔보세요.
  plt.text(i, value, str(value), ha='center', va='bottom', color='black', fontsize=10)

plt.title('Bar Chart with Text Annotations')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### **수평 막대 그래프 : 성별에 따른 생존자 수 확인하기**

슬라이드20.JPG

#### 프롬프트
> pandas를 사용해서 ‘Survived’ 컬럼이 1인 데이터 중 ‘Sex’ 컬럼별 생존자 수를 계산하는 코드를 작성해줘.

> pandas 데이터프레임 ‘titanic’에서 ‘Survived’ 값이 1인 행들만 필터링한 뒤, ‘Sex’ 컬럼을 기준으로 생존자 수를 value_counts()로 집계하는 코드를 작성해줘.

```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 성별에 따른 생존자의 수 계산
survived_counts = titanic[titanic['Survived'] == 1]['Sex'].value_counts()
survived_counts
```

#### 프롬프트
> matplotlib.pyplot을 사용해 ‘survived_counts’ 데이터를 기반으로 성별 생존자 수를 나타내는 수평 막대 그래프를 그리고, 그래프에 제목, 축 레이블, 범례를 추가하며, 생존자 수를 막대 끝에 표시하는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 사용해서 ‘survived_counts’ 변수의 인덱스(성별)와 값(생존자 수)으로 수평 막대 그래프(barh)를 그려, 막대 색상은 여성은 ‘darkturquoise’, 남성은 ‘salmon’으로 지정하고, 그래프 제목은 ‘Survived Counts by Gender on Titanic’, x축 레이블은 ‘Count’, y축 레이블은 ‘Gender’로 설정하며, 범례는 ‘Survived - Female’, ‘Survived - Male’를 오른쪽 위에 표시해. 또한 남성 생존자 수 위치에 회색 점선 수평선을 추가하고, 각 막대 끝에 생존자 수 값을 막대 오른쪽에 텍스트로 표시하는 코드를 작성해줘.

```python
import matplotlib.pyplot as plt

# 수평 막대 그래프 그리기
bars = plt.barh(survived_counts.index, survived_counts, color=['darkturquoise', 'salmon'])
plt.title('Survived Counts by Gender on Titanic')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.legend(bars, ['Survived - Female', 'Survived - Male'], loc='upper right')

# 차이 강조를 위해 수평선 추가
plt.axvline(x=survived_counts['male'], color='gray', linestyle='--', linewidth=1)

# 생존자 수 표시
for i, value in enumerate(survived_counts):
  plt.text(value + 1, i, str(value), ha='left', va='center')

plt.show()
```

### **파이 차트 : 생존자, 사망자 비율 표현하기**

슬라이드22.JPG

#### 프롬프트
> pandas를 사용해 ‘Survived’ 컬럼에서 생존자와 사망자의 수를 계산하는 코드를 작성해줘.

> pandas 데이터프레임 ‘titanic’에서 ‘Survived’ 컬럼의 각 값(0과 1)에 대해 value_counts() 함수를 사용해 사망자(0)와 생존자(1)의 수를 집계하는 코드를 작성해줘.



```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 사망자와 생존자의 수 계산
survived_counts = titanic['Survived'].value_counts()
survived_counts
```

#### 프롬프트

> 파이썬 matplotlib.pyplot 라이브러리를 불러와 ‘survived_counts’ 데이터를 이용해 파이 차트를 그리고, 각 조각에 ‘Not Survived’와 ‘Survived’ 라벨을 붙이며, 색상은 주황색과 금색으로 설정하고, 퍼센트는 소수점 첫째 자리까지 표시하며, 시작 각도는 90도, 그림자 효과와 생존자 조각만 0.1만큼 돌출(폭발)시켜서 ‘Survival Distribution on the Titanic’ 제목과 함께 표시하는 코드를 작성해줘.



```python
import matplotlib.pyplot as plt

# 파이 차트 그리기
plt.figure(figsize=(8, 8))
plt.pie(survived_counts, labels=['Not Survived', 'Survived'], colors=['orange', 'gold'],
        autopct='%0.1f%%', startangle=90, shadow=True, explode=(0, 0.1))

plt.title('Survival Distribution on the Titanic')
plt.show()
```

#### 프롬프트
> pandas를 사용해서 ‘titanic_selena.csv’ 파일을 읽고, ‘Age’ 컬럼에 결측치가 있는 행을 제거한 뒤, 처리 전과 후의 데이터프레임 정보를 출력하는 코드를 작성해줘.

> 파이썬 pandas 라이브러리를 불러와 ‘titanic_selena.csv’ 파일을 read_csv() 함수로 읽어 ‘titanic’ 변수에 저장하고, info() 메서드로 처리 전 데이터프레임 정보를 출력한 후, ‘Age’ 컬럼에 결측치가 있는 행을 dropna(subset=['Age'])로 제거하고, 다시 info() 메서드로 처리 후 데이터프레임 정보를 출력하는 코드를 작성해줘.

### **히트맵 : 두 변수의 상관 관계를 표시하기**

슬라이드25.JPG

#### 프롬프트
> pandas를 사용해서 ‘Age’와 ‘Fare’ 컬럼의 결측치를 제거하고, ‘PassengerId’ 열을 제외한 나머지 수치형 열들에 대해 상관 행렬을 계산하는 코드를 작성해줘.

> pandas 데이터프레임 ‘titanic’에서 ‘Age’와 ‘Fare’ 컬럼에 결측치가 있는 행을 dropna(subset=['Age', 'Fare'])로 제거한 뒤, ‘PassengerId’ 열을 drop()으로 삭제하고, 나머지 수치형 컬럼들에 대해 corr(numeric_only=True) 함수를 사용해 상관 행렬을 계산하는 코드를 작성해줘.



```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 결측치 처리
titanic = titanic.dropna(subset=['Age', 'Fare'])

# 상관 행렬 계산
correlation_matrix = titanic.drop('PassengerId', axis=1).corr(numeric_only=True)
correlation_matrix
```

```python
list(range(len(correlation_matrix.columns)))
```

#### 프롬프트
> matplotlib.pyplot을 사용해 상관 행렬 데이터를 히트맵 형태로 시각화하는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 불러와서 ‘correlation_matrix’ 데이터를 ‘PuRd_r’ 컬러맵을 사용해 matshow() 함수로 히트맵을 그리고, 컬러바를 추가하며, x축과 y축 눈금에 상관 행렬 컬럼 이름을 각각 45도 회전해서 표시하고, 제목은 ‘Correlation Heatmap of Titanic’으로 설정하는 코드를 작성해줘.

```python
import matplotlib.pyplot as plt

# 히트맵 그리기
plt.matshow(correlation_matrix, cmap='PuRd_r')
plt.colorbar()

# x축과 y축의 눈금 설정
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title('Correlation Heatmap of Titanic')
plt.show()
```

### **타이타닉 데이터셋으로 개별 서브플롯 하나씩 그리기**

슬라이드35.JPG

* Survived가 0이면 사망자를, 1이면 생존자를 나타냄

#### 프롬프트
> pandas를 사용해 ‘Parch’별로 ‘Survived’ 값에 따른 생존자와 사망자 수를 계산하는 코드를 작성해줘.



> pandas 데이터프레임 ‘titanic’에서 ‘Parch’ 컬럼을 기준으로 그룹화한 뒤, ‘Survived’ 컬럼의 값별 개수를 value_counts()로 집계하고, unstack()으로 피벗해 결측치는 fillna(0)로 0으로 채우는 코드를 작성해줘.

```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 각 부모와 자녀의 수에 따른 생존자와 사망자 수 계산
parch_counts = titanic.groupby('Parch')['Survived'].value_counts().unstack().fillna(0)
parch_counts
```

```python
parch_counts[0].values
```

#### 프롬프트
> matplotlib.pyplot을 사용해 ‘Parch’ 값별로 사망자 수는 선 그래프로, 생존자 수는 막대 그래프로 나타내는 두 개의 세로 서브플롯을 그리는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 사용해서 parch_counts 데이터프레임의 인덱스를 문자열로 변환한 x축과, 사망자 수(y1)는 보라색(indigo)의 선 그래프(점 표시 포함), 생존자 수(y2)는 진분홍색(deeppink)의 막대 그래프로 각각 그리고, 각 서브플롯에 x축 레이블 ‘Parch’, y축 레이블과 색상을 지정하며, 범례를 오른쪽 위에 표시하고, 전체 그래프에는 ‘Survival Analysis by Number of Parents/Children (Parch) on the Titanic’ 제목을 슈퍼 타이틀로 넣은 후, tight_layout()으로 간격 조절하고 그래프를 출력하는 코드를 작성해줘.

```python
import matplotlib.pyplot as plt

# x, y 설정
x = parch_counts.index.astype(str) # 데이터프레임의 인덱스를 문자열로 반환
y1 = parch_counts[0].values # 사망자
y2 = parch_counts[1].values # 생존자

plt.figure(figsize=(10, 10))

# plt.subplot() 함수를 사용하여 하나의 그림인 figure에 개별 서브플롯을 하나씩 추가
# 첫 번째 서브플롯 설정(선 그래프)
plt.subplot(2, 1, 1)
# 선 그래프 그리기
plt.plot(x, y1, '-o', color='indigo', markersize=7, linewidth=3, alpha=0.7,
label='Not Survived')
plt.xlabel('Parch')
plt.ylabel('Not Survived Count', color='indigo')
plt.tick_params(axis='y', labelcolor='indigo')
plt.legend(loc='upper right')

# 두 번째 서브플롯 설정(막대 그래프)
plt.subplot(2, 1, 2)
plt.bar(x, y2, color='deeppink', alpha=0.7, width=0.5, label='Survived')
plt.xlabel('Parch')
plt.ylabel('Survived Count', color='deeppink')
plt.tick_params(axis='y', labelcolor='deeppink')
plt.legend(loc='upper right')

# 제목 설정
plt.suptitle('Survival Analysis by Number of Parents/Children (Parch) on the Titanic')
plt.tight_layout()
plt.show()
```

### **하나의 서브플롯에 여러 그래프 그리기**

슬라이드37.JPG

#### 프롬프트
> pandas를 사용해 ‘Parch’별 ‘Survived’ 상태에 따른 생존자와 사망자 수를 계산하는 코드를 작성해줘.



> pandas 데이터프레임 ‘titanic’에서 ‘Parch’ 컬럼을 기준으로 그룹화한 후, ‘Survived’ 컬럼의 값별 개수를 value_counts()로 집계하고, unstack()으로 행렬 형태로 변환하며, 결측치는 fillna(0)으로 0으로 채우는 코드를 작성해줘.

```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 각 부모와 자녀의 수에 따른 생존자와 사망자 수 계산
parch_counts = titanic.groupby('Parch')['Survived'].value_counts().unstack().fillna(0)
parch_counts
```

#### 프롬프트
> matplotlib.pyplot을 사용해 ‘Parch’별 사망자 수를 선 그래프로, 생존자 수를 같은 x축에 두 번째 y축을 가진 막대 그래프로 그리는 코드를 작성해줘.

> 파이썬 matplotlib.pyplot 라이브러리를 사용해 parch_counts 데이터의 인덱스를 문자열로 변환한 x축 값에 대해, 첫 번째 y축에는 보라색(indigo) 선 그래프(마커는 정사각형, 두께 5, 투명도 0.7)로 사망자 수를 그리고, 두 번째 y축을 twinx()로 생성해 진분홍색(deeppink) 막대 그래프로 생존자 수를 나타내며, 각 축에 레이블과 눈금 색상을 맞추고, 범례는 오른쪽 위에 약간 위치를 달리해 표시하고, 전체 제목은 ‘Survival Analysis by Number of Parents/Children (Parch) on the Titanic’로 설정한 후 tight_layout()으로 간격을 조절하고 그래프를 출력하는 코드를 작성해줘.

```python
import matplotlib.pyplot as plt

# x, y값 설정
x = parch_counts.index.astype(str) # 데이터프레임의 인덱스를 문자열로 변환
y1 = parch_counts[0].values # 사망자
y2 = parch_counts[1].values # 생존자

# plt.subplots() 함수를 사용하여 그래프 생성
# figure와 서브플롯 생성
fig, ax1 = plt.subplots()

# 첫 번째 그래프 설정 (선 그래프)
ax1.plot(x, y1, '-s', color='indigo', markersize=7, linewidth=5, alpha=0.7,
         label='Not Survived')
ax1.set_xlabel('Parch') # x축의 레이블
ax1.set_ylabel('Not Survived Count', color='indigo') # y축의 레이블
ax1.tick_params(axis='y', labelcolor='indigo') # 눈금의 모양과 속성
ax1.legend(loc='upper right', bbox_to_anchor=(1, 1)) # 범례

# .twinx() 함수를 사용하여 두 번째 그래프와 축을 공유하는 새로운 Axes 객체 생성
ax2 = ax1.twinx()

# 두 번째 그래프 설정 (막대 그래프)
ax2.bar(x, y2, color='deeppink', alpha=0.7, width=0.7, label='Survived')
ax2.set_ylabel('Survived Count', color='deeppink') # 서브플롯 y축에 레이블 추가
ax2.tick_params(axis='y', labelcolor='deeppink')
ax2.legend(loc='upper right', bbox_to_anchor=(1, 0.9))

plt.suptitle('Survival Analysis by Number of Parents/Children (Parch) on the Titanic')
plt.tight_layout()
plt.show()
```
